```python
# Install Streamlit if not already installed
try:
    import streamlit
except ImportError:
    !pip install streamlit -qq
    import streamlit

import streamlit as st
import pandas as pd
import json
import os
import plotly.express as px
from openai import OpenAI


# -----------------------------------------------------------------------------
# CONFIGURAÇÃO DE PÁGINA E ESTILO UX
# -----------------------------------------------------------------------------
st.set_page_config(
    page_title="FinAI — Agente Financeiro Inteligente",
    page_icon="🤖",
    layout="wide"
)

st.markdown("""
    <style>
    .disclaimer-box {
        background-color: #1e293b;
        border-left: 4px solid #f59e0b;
        padding: 10px 14px;
        border-radius: 6px;
        font-size: 0.85rem;
        color: #cbd5e1;
        margin-bottom: 20px;
    }
    </style>
""", unsafe_allow_html=True)

# -----------------------------------------------------------------------------
# CARREGAMENTO DOS DADOS MOCKADOS (data/)
# -----------------------------------------------------------------------------
@st.cache_data
def carregar_dados():
    # Use os.getcwd() as base_dir since __file__ is not defined in Colab notebooks
    base_dir = os.getcwd()
    data_dir = os.path.join(base_dir, "data")

    # Create 'data' directory if it doesn't exist, to prevent errors for missing directory
    if not os.path.exists(data_dir):
        os.makedirs(data_dir, exist_ok=True)

    # Defaults de fallback caso a pasta data ainda esteja vazia
    perfil = {"nome": "Cliente Exemplo", "perfil": "Conservador", "objetivo": "Reserva de Emergência"}
    produtos = [{"nome": "Tesouro Selic", "categoria": "Renda Fixa", "risco": "Baixo"}]
    transacoes = pd.DataFrame([{"data": "2026-01-01", "categoria": "Alimentacao", "valor": 150.0}])
    atendimentos = pd.DataFrame([{"data": "2026-01-01", "assunto": "Dúvida sobre taxa"}])

    perfil_path = os.path.join(data_dir, "perfil_investidor.json")
    if os.path.exists(perfil_path):
        with open(perfil_path, "r", encoding="utf-8") as f:
            perfil = json.load(f)

    produtos_path = os.path.join(data_dir, "produtos_financeiros.json")
    if os.path.exists(produtos_path):
        with open(produtos_path, "r", encoding="utf-8") as f:
            produtos = json.load(f)

    transacoes_path = os.path.join(data_dir, "transacoes.csv")
    if os.path.exists(transacoes_path):
        transacoes = pd.read_csv(transacoes_path)

    atendimentos_path = os.path.join(data_dir, "historico_atendimento.csv")
    if os.path.exists(atendimentos_path):
        atendimentos = pd.read_csv(atendimentos_path)

    return perfil, produtos, transacoes, atendimentos

perfil, produtos, df_transacoes, df_atendimentos = carregar_dados()

# -----------------------------------------------------------------------------
# PERSISTÊNCIA DE SESSÃO
# -----------------------------------------------------------------------------
if "messages" not in st.session_state:
    st.session_state.messages = [
        {"role": "assistant", "content": f"Olá! Sou o **FinAI**, seu agente financeiro consultivo. Analisei seus dados mais recentes e estou pronto para ajudar a planejar suas metas."}
    ]

# -----------------------------------------------------------------------------
# SIDEBAR - CONTEXTO E CONFIGURAÇÕES
# -----------------------------------------------------------------------------
with st.sidebar:
    st.title("⚙️ Configurações do Agente")
    api_key = st.text_input("OpenAI API Key", type="password")

    st.divider()
    st.subheader("👤 Perfil do Cliente (data/)")
    st.json(perfil)

    if st.button("🧹 Limpar Histórico"):
        st.session_state.messages = [st.session_state.messages[0]]
        st.rerun()

# -----------------------------------------------------------------------------
# CORPO PRINCIPAL & NAVEGAÇÃO
# -----------------------------------------------------------------------------
st.title("🤖 Agente Financeiro Inteligente (FinAI)")

st.markdown("""
<div class="disclaimer-box">
    🛡️ <b>Mecanismo Anti-Alucinação Ativo:</b> As recomendações do agente são baseadas estritamente na base de conhecimento cadastrada e nos dados de perfil do cliente.
</div>
""", unsafe_allow_html=True)

tab_chat, tab_dados, tab_simulador = st.tabs(["💬 Chat Consultivo", "📊 Base de Conhecimento", "📈 Simulações"])

# --- TAB 1: CHAT CONVERSACIONAL ---
with tab_chat:
    for msg in st.session_state.messages:
        with st.chat_message(msg["role"]):
            st.markdown(msg["content"])

    if user_input := st.chat_input("Ex: Analise meus últimos gastos ou sugira um produto adequado ao meu perfil."):
        st.session_state.messages.append({"role": "user", "content": user_input})
        with st.chat_message("user"):
            st.markdown(user_input)

        with st.chat_message("assistant"):
            if api_key:
                try:
                    client = OpenAI(api_key=api_key)
                    system_prompt = f"""
                    Você é um agente financeiro proativo, seguro e consultivo.
                    Contexto do Cliente: {json.dumps(perfil)}
                    Produtos Disponíveis: {json.dumps(produtos)}

                    Regras Rígidas de Segurança:
                    1. NUNCA prometa rentabilidade futura.
                    2. Responda baseado apenas nos produtos disponíveis.
                    3. Mantenha um tom profissional, didático e empático.
                    """

                    full_messages = [{"role": "system", "content": system_prompt}] + st.session_state.messages
                    res = client.chat.completions.create(model="gpt-4o-mini", messages=full_messages)
                    reply = res.choices[0].message.content
                except Exception as e:
                    reply = f"Erro ao conectar à API: {str(e)}"
            else:
                reply = f"*(Modo Demonstrativo)* Analisando seu perfil **{perfil.get('perfil', 'Conservador')}**, recomendo analisar os produtos de renda fixa disponíveis no catálogo para sua meta de {perfil.get('objetivo', 'Reserva de Emergência')}."

            st.markdown(reply)
            st.session_state.messages.append({"role": "assistant", "content": reply})

# --- TAB 2: VISUALIZAÇÃO DOS DADOS MOCKADOS ---
with tab_dados:
    st.subheader("📁 Dados Integrados da Pasta `data/`")
    c1, c2 = st.columns(2)
    with c1:
        st.markdown("**Histórico de Transações (`transacoes.csv`)**")
        st.dataframe(df_transacoes, use_container_width=True)
    with c2:
        st.markdown("**Produtos Financeiros (`produtos_financeiros.json`)**")
        st.json(produtos)

# --- TAB 3: CÁLCULOS DETERMINÍSTICOS ---
with tab_simulador:
    st.subheader("🧮 Calculadora com Garantia Anti-Alucinação")

    val_inicial = st.number_input("Valor Inicial (R$)", value=1000.0)
    val_mensal = st.number_input("Aporte Mensal (R$)", value=200.0)
    meses = st.slider("Prazo em Meses", 1, 60, 12)

    # Cálculo executado exclusivamente em Python
    total_investido = val_inicial + (val_mensal * meses)
    saldo_estimado = total_investido * 1.08  # Exemplo simulação 8% a.a.

    st.metric("Total Investido", f"R$ {total_investido:,.2f}")
    st.metric("Saldo Final Estimado", f"R$ {saldo_estimado:,.2f}")
```
